## 多线程获取关注特征

在已经获得基本信息的情况下，添加 followers 对 **influencers** 关注情况
> 也即这些 followers 对 `{Taylor Swift, ESPN, Bill Gates, YouTube, CNN, Barack Obama, Kim Kardashian, Cristiano Ronaldo, Jimmy Fallon, Oprah Winfrey, Lil Wayne, NASA}` 这 12 个用户的关注情况

### 获取基本特征和所有 Tokens

````python
import pandas as pd
import tweepy
import threading
import numpy as np

# 获取基本特征 
basic = pd.read_csv('./basic_feas.csv')
# 获取所有账号
all_tokens = getTokens()
````

### `basic` 表分块
对 `basic` 进行分块，一个分为 34 块，存到 `blocks` 中（深拷贝）

````python
start = 0
stop = 10000
step = 300

blocks = []
for i in range(start, stop, step):
    blocks.append(basic[i:min(i+step, stop)].copy(deep=True))
````

````python
len(blocks)
````

### 定义关注特征必要变量以及函数
12 个 `influencer` 的 `screen_name` 定义 + `influencer` 特征获取必要的函数定义：`isAfollowingB` 和 `get_influencers_info`

````python
# 12 个 influencer 的 screen_name
u_TaylorSwift = 'taylorswift13'
u_ESPN = 'espn'
u_BillGates = 'BillGates'
u_YouTube = 'YouTube'
u_CNN = 'CNN'
u_BarackObama = 'BarackObama'
u_KimKardashian = 'KimKardashian'
u_CristianoRonaldo = 'Cristiano'
u_JimmyFallon = 'jimmyfallon'
u_OprahWinfrey = 'Oprah'
u_LilWayne = 'LilTunechi'
u_NASA = 'NASA'

def isAfollowingB(api, A, B):
    # 查看是否 A follow 了 B（A 和 B 都是 screen_name）
    status = api.show_friendship(source_screen_name=A, target_screen_name=B)
    return 1 if status[0].following else 0

def get_influencers_info(api, user):
    # user 是 User Object
    if user.friends_count >= 0:
        u_TaylorSwift_following = isAfollowingB(api, user.screen_name, u_TaylorSwift)
        u_ESPN_following = isAfollowingB(api, user.screen_name, u_ESPN)
        u_BillGates_following = isAfollowingB(api, user.screen_name, u_BillGates)
        u_YouTube_following = isAfollowingB(api, user.screen_name, u_YouTube)
        u_CNN_following = isAfollowingB(api, user.screen_name, u_CNN)
        u_BarackObama_following = isAfollowingB(api, user.screen_name, u_BarackObama)
        u_KimKardashian_following = isAfollowingB(api, user.screen_name, u_KimKardashian)
        u_CristianoRonaldo_following = isAfollowingB(api, user.screen_name, u_CristianoRonaldo)
        u_JimmyFallon_following = isAfollowingB(api, user.screen_name, u_JimmyFallon)
        u_OprahWinfrey_following = isAfollowingB(api, user.screen_name, u_OprahWinfrey)
        u_LilWayne_following = isAfollowingB(api, user.screen_name, u_LilWayne)
        u_NASA_following = isAfollowingB(api, user.screen_name, u_NASA)
        
        return [ u_TaylorSwift_following, u_ESPN_following, u_BillGates_following, u_YouTube_following, 
            u_CNN_following, u_BarackObama_following, u_KimKardashian_following, u_CristianoRonaldo_following, 
            u_JimmyFallon_following, u_OprahWinfrey_following, u_LilWayne_following, u_NASA_following ]
    
    else:
        f_list = []
        for friend in tweepy.Cursor(api.friends, screen_name=user.screen_name, count=200).items():
            f_list.append(friend.screen_name)

        u_TaylorSwift_following = 1 if u_TaylorSwift in f_list else 0
        u_ESPN_following = 1 if u_ESPN in f_list else 0
        u_BillGates_following = 1 if u_BillGates in f_list else 0
        u_YouTube_following = 1 if u_YouTube in f_list else 0
        u_CNN_following = 1 if u_CNN in f_list else 0
        u_BarackObama_following = 1 if u_BarackObama in f_list else 0
        u_KimKardashian_following = 1 if u_KimKardashian in f_list else 0
        u_CristianoRonaldo_following = 1 if u_CristianoRonaldo in f_list else 0
        u_JimmyFallon_following = 1 if u_JimmyFallon in f_list else 0
        u_OprahWinfrey_following = 1 if u_OprahWinfrey in f_list else 0
        u_LilWayne_following = 1 if u_LilWayne in f_list else 0
        u_NASA_following = 1 if u_NASA in f_list else 0

        return [ u_TaylorSwift_following, u_ESPN_following, u_BillGates_following, u_YouTube_following, 
                u_CNN_following, u_BarackObama_following, u_KimKardashian_following, u_CristianoRonaldo_following, 
                u_JimmyFallon_following, u_OprahWinfrey_following, u_LilWayne_following, u_NASA_following ]
````

### 多线程进行 followers 对 influencer 的关注情况的添加
> 34 块基本特征表，就需要 34 个开发者账号，

````python
def add_influ_feas_to_block(k):
    '''
    k: 表示第 k 个特征表，同样就使用第 k 个开发者账号
    '''
    print('{} : Processing block {}'.format(threading.current_thread().name, k))
    auth = tweepy.OAuthHandler(all_tokens[k]['consumer_key'], all_tokens[k]['consumer_secret'])
    auth.set_access_token(all_tokens[k]['access_token'], all_tokens[k]['access_token_secret'])

    api = tweepy.API(auth, wait_on_rate_limit=True)
            
    for i in blocks[k].index:
        try:
            print(' *{}* '.format(i), end='\t')
            u = api.get_user(screen_name = blocks[k].loc[i, 'screen_name'])
            ret = get_influencers_info(api, u)
            blocks[k].loc[i, 'u_TaylorSwift_following'] = ret[0]
            blocks[k].loc[i, 'u_ESPN_following'] = ret[1]
            blocks[k].loc[i, 'u_BillGates_following'] = ret[2]
            blocks[k].loc[i, 'u_YouTube_following'] = ret[3]
            blocks[k].loc[i, 'u_CNN_following'] = ret[4]
            blocks[k].loc[i, 'u_BarackObama_following'] = ret[5]
            blocks[k].loc[i, 'u_KimKardashian_following'] = ret[6]
            blocks[k].loc[i, 'u_CristianoRonaldo_following'] = ret[7]
            blocks[k].loc[i, 'u_JimmyFallon_following'] = ret[8]
            blocks[k].loc[i, 'u_OprahWinfrey_following'] = ret[9]
            blocks[k].loc[i, 'u_LilWayne_following'] = ret[10]
            blocks[k].loc[i, 'u_NASA_following'] = ret[11]
        except Exception as e:
            print('{} :{} {} for block {}'.format(threading.current_thread().name, e, i, k))
        
    
    print('{} : Done for block {}'.format(threading.current_thread().name, k))
````

````python
for i in range(0, len(blocks)):
    t = threading.Thread(target=add_influ_feas_to_block, args=(i,), name='ForBlock-'+str(i))
    t.start()
````

### 重复执行 1.5 步以尽可能爬取所有未注销用户
绝大多数 followers 在处理后应该都已经具有了关注特征，但依然存在一些 follower 其关注特征全为 `NAN`

一般有两种原因：

* 获取关注特征时，翻墙网络问题导致获取失败（超时）：可处理， `np.isnan(blocks[k].loc[i, 'u_TaylorSwift_following'])` 再收集多次
> ```shell
ForBlock-12 :Failed to send request: HTTPSConnectionPool(host='api.twitter.com', port=443): Max retries exceeded with url: /1.1/friendships/show.json?source_screen_name=nogi_46soccer&target_screen_name=espn (Caused by ProxyError('Cannot connect to proxy.', NewConnectionError('<urllib3.connection.VerifiedHTTPSConnection object at 0x000001475BA0D860>: Failed to establish a new connection: [WinError 10048] 通常每个套接字地址(协议/网络地址/端口)只允许使用一次。',))) 3882 for block 12
```

* 在第一步收集的 10000 个用户中，有用户注销了 twitter 账号：不可处理，只能再额外收集新的用户
> ```shell
ForBlock-22 :[{'code': 50, 'message': 'User not found.'}] 6872 for block 22
```

````python
def add_influ_feas_to_block_2(k):
    '''
    k: 表示第 k 个特征表，同样就使用第 k 个开发者账号
    '''
    print('{} : Processing block {}'.format(threading.current_thread().name, k))
    auth = tweepy.OAuthHandler(all_tokens[k]['consumer_key'], all_tokens[k]['consumer_secret'])
    auth.set_access_token(all_tokens[k]['access_token'], all_tokens[k]['access_token_secret'])

    api = tweepy.API(auth, wait_on_rate_limit=True)
            
    for i in blocks[k].index:
        try:
            # 仅处理由于代理错误而没有补充关注信息的 followers
            if np.isnan(blocks[k].loc[i, 'u_TaylorSwift_following']):
                print(' *{}* '.format(i), end='\t')
                u = api.get_user(screen_name = blocks[k].loc[i, 'screen_name'])
                ret = get_influencers_info(api, u)
                blocks[k].loc[i, 'u_TaylorSwift_following'] = ret[0]
                blocks[k].loc[i, 'u_ESPN_following'] = ret[1]
                blocks[k].loc[i, 'u_BillGates_following'] = ret[2]
                blocks[k].loc[i, 'u_YouTube_following'] = ret[3]
                blocks[k].loc[i, 'u_CNN_following'] = ret[4]
                blocks[k].loc[i, 'u_BarackObama_following'] = ret[5]
                blocks[k].loc[i, 'u_KimKardashian_following'] = ret[6]
                blocks[k].loc[i, 'u_CristianoRonaldo_following'] = ret[7]
                blocks[k].loc[i, 'u_JimmyFallon_following'] = ret[8]
                blocks[k].loc[i, 'u_OprahWinfrey_following'] = ret[9]
                blocks[k].loc[i, 'u_LilWayne_following'] = ret[10]
                blocks[k].loc[i, 'u_NASA_following'] = ret[11]
        except Exception as e:
            print('{} :{} {} for block {}'.format(threading.current_thread().name, e, i, k))
        
    
    print('{} : Done for block {}'.format(threading.current_thread().name, k))
````

````python
for i in range(0, len(blocks)):
    t = threading.Thread(target=add_influ_feas_to_block_2, args=(i,), name='ForBlock-'+str(i))
    t.start()
````
